```python
road_gdf = edge_gdf[edge_gdf['maxspeed'].notna()]
```

```python
road_gdf['osmid'] = pd.to_numeric(road_gdf['osmid'], errors='coerce')
road_gdf.to_file('..\\output\\road_gdf.GEOJSON')
```

```python

for col in road_gdf.columns:
    road_gdf[col] = road_gdf[col].apply(lambda x: x[0] if type(x) == list else x)
road_gdf.to_file('..\\output\\road_gdf.GEOJSON')
```
